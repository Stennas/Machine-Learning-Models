# K-Means process step by step using your synthetic data

## âœ… Step 0: Generate synthetic data

```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate 700 data points, 4 centers, a bit of noise
X, _ = make_blobs(n_samples=700, centers=4, cluster_std=2.3, random_state=42)

# Plot the raw data (before clustering)
plt.scatter(X[:, 0], X[:, 1], s=30, alpha=0.6)
plt.title("Synthetic Data (Before Clustering)")
plt.show()
```

## âœ… Step 1: Choose the number of clusters (K)

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=42)
```

## âœ… Step 2: Fit the K-Means model

```python
kmeans.fit(X)
```

This does all the steps:

- Initializes centroids

- Assigns points

- Recalculates centroids

- Repeats until convergence

## âœ… Step 3: Check predictions (cluster assignments)

```python
labels = kmeans.labels_
```

## âœ… Step 4: Visualize the clusters

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=30, alpha=0.7)
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=100, marker='X')  # plot centroids
plt.title("K-Means Clustering (K=4)")
plt.show()
```

```python
print("Iterations until convergence:", kmeans.n_iter_)
```
